```python
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
IMG=cv2.imread('original.jpg')
```

```python
plt.imshow(IMG)
plt.show()
```

```python
IMG = cv2.cvtColor(IMG, cv2.COLOR_BGR2RGB)
plt.imshow(IMG)
plt.show()
```

```python
# file='savedimage.JPG'
# cv2.imwrite(file, IMG)
IMG.shape
(4650, 4650, 3)
vect_img=IMG.reshape(IMG.shape[0]*IMG.shape[1], IMG.shape[2])
vect_img.shape
```

```python
kmeans = KMeans(n_clusters=6, random_state=0).fit(vect_img)
pic2show = kmeans.cluster_centers_[kmeans.labels_]
```

```python

 
np.savetxt('data.csv', pic2show, delimiter=',')
# loading the data
#data = loadtxt('data.csv', delimiter=',')
 
```

```python
pic2show.shape
```

```python
#restored image is the final output image
restored=pic2show.reshape(IMG.shape[0], IMG.shape[1], IMG.shape[2])
restored=restored*255
plt.imshow((restored).astype(np.uint8))
plt.show()
```
